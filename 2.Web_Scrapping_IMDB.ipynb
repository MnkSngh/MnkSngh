# Importing Libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
import openpyxl as pl
```

# Creating Excel Workbook to write scraped Data to Sheet.

```python
excel = pl.Workbook()
```

```python
sheet = excel.active
```

```python
sheet.title= "Movie Ratings"
```

```python
sheet.append(['Rank of the movie', 'Name of the movie', 'Release Year', 'Rating'])
```

# Starting Web Scraping

```python
url = "https://www.imdb.com/chart/top/" #url of the site.
```

```python
source = requests.get(url, verify = False) # getting the url to python
```

```python
soup = BeautifulSoup(source.text, 'html.parser') #parsing the website source code to Html
```

```python
movies = soup.find('tbody', class_ = "lister-list") # finding specific tag from Html code to scrape Data
```

```python
listss = movies.find_all('tr')
listss
```

```python
#looping through all the tags containings our required text and appending it to the excel sheet created above 
for lists in listss:
    name =   lists.find('td', class_ = "titleColumn").a.text
    rank =   lists.find('td', class_ = "titleColumn").get_text(strip = True).split('.')[0]
    year =   lists.find('td', class_ = "titleColumn").span.text.strip('()')
    rating = lists.find('td', class_ = "ratingColumn imdbRating").strong.text
    print(rank, name, year, rating)
    sheet.append([rank, name, year, rating])  
```

# Saving the excel workbook after data got written to the sheet.

```python
excel.save('Web_scraping 5 of IMDB.xlsx') #Saving the excel workbook after data got written to the sheet.
```
