```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://www.scrapethissite.com/pages/ajax-javascript/#2015'
driver.get(url)
time.sleep(5) ## Required to extract the whole HTML correctly
html = driver.execute_script("return document.documentElement.outerHTML")##alternative way to get the page HTML content so 
                                                                        ##that we can prase it later using Beautifulsoup
```

```python
html
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```
